```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
cov=pd.read_csv('national-history.csv')
comments=pd.read_csv('comments_byDate.csv')
```

```python
comments.date = pd.to_datetime(comments.date)
```

```python
f, axs = plt.subplots(1,1,figsize=(12,8))
plt.scatter('date', 'count', data=comments,s=5,alpha = 0.5)
plt.plot('date', 'count', data=comments)
plt.title("How Many Comments We get for Each Day")
plt.xlabel("Date")
plt.ylabel("Count")
```

```python
f, axs = plt.subplots(1,1)
sns.boxplot(x="count",data=comments)
sns.swarmplot(x="count", data=comments, color=".25",alpha = .5)
axs.set_title('Reddit Comments Counts')
```

```python
cov.date = pd.to_datetime(cov.date)
```

```python
f, axs = plt.subplots(1,1,figsize=(8,6))
plt.plot('date', 'positive', data=cov,color = 'orange')
plt.title("How Does COV-19 Positive Cases Amounts Change by Time")
plt.xlabel("Date")
plt.ylabel("Positive Cases Counts (*10^7)")
plt.show()
```

```python
cov1= cov
```

```python
cov1.death = cov1.death/1000
```

```python
cov1.inIcuCumulative = cov1.inIcuCumulative/1000
```

```python
cov1.hospitalizedCumulative = cov1.hospitalizedCumulative/1000
```

```python
cov1.date = pd.to_datetime(cov1.date)
```

```python
f, axs = plt.subplots(1,1,figsize=(8,6))
plt.plot('date', 'death', data=cov1,color = 'blue')
plt.plot('date', 'inIcuCumulative', data=cov1,color = 'green')
plt.plot('date', 'hospitalizedCumulative', data=cov1,color = 'red')
plt.title("How Does death rate and Hospitalied Data Changed by Time")
plt.xlabel("Date")
plt.ylabel("Counts(*1000)")
plt.legend()
plt.show()
```

```python
f, ax = plt.subplots(figsize=(10, 8))
corr = cov.corr()
#cmap=sns.diverging_palette(220, 10, as_cmap=True)
cmap=sns.cubehelix_palette(9, start=2, rot=0, dark=0, light=.95,as_cmap=True)
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=cmap,
            square=True, ax=ax)
ax.set_title("Correlations of Features of COV-19 Data")
```
